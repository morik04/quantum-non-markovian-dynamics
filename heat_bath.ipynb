```python
import numpy as np
import math
omega_c=1.5
M=60
zeta=1
mj=1
h=1
```

```python
omega_j=[]
for j in range(1, M+1):
    omega_j.append(omega_c*math.log(M/(M-j+0.5)))
print(omega_j)
```

```python
cj=[]
for j in range(1, M+1):
    cj.append(M**(-1/2)*math.sqrt(zeta*mj*h*omega_c)*omega_j[j-1])
print(cj)
```

```python
omega=np.linspace(0, 8, 500)
J=(np.pi/2)*h*zeta*omega*np.exp(-1*omega/omega_c)

plt.plot(omega, J)
plt.xlabel('Omega')
plt.ylabel('J(omega)')
plt.title('Spectral Density Function')
plt.grid()
plt.show()
```
