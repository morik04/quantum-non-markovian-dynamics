Hamiltonian Evolution

```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.linalg import expm
```

```python
delta_t=0.25
t=np.arange(0, 10, 0.01)
t_trot=np.arange(0, 10, delta_t)
```

```python
J=1  #coupling strength
X=np.array([[0, 1], [1, 0]])
Y=np.array(([0, -1j], [1j, 0]))
input_v=np.array([[0], [1], [0], [0]])
hamiltonian_X=J*0.5*np.kron(X, X)
hamiltonian_Y=J*0.5*np.kron(Y, Y)
hamiltonian=hamiltonian_X+hamiltonian_Y
```

```python
print("Hamiltonian:\n", hamiltonian)
```

```python
a_real=[]
a_imag=[]

b_real=[]
b_imag=[]

c_real=[]
c_imag=[]

d_real=[]
d_imag=[]
```

```python
for i in range(len(t)):
    U=expm(-1j*hamiltonian*t[i])@input_v
    a_real.append(np.real(U[0, 0]))
    a_imag.append(np.imag(U[0, 0]))
    
    b_real.append(np.real(U[1, 0]))
    b_imag.append(np.imag(U[1, 0]))
    
    c_real.append(np.real(U[2, 0]))
    c_imag.append(np.imag(U[2, 0]))
    
    d_real.append(np.real(U[3, 0]))
    d_imag.append(np.imag(U[3, 0]))
```

```python
print("a_real:", a_real)
print("a_imag:", a_imag)
print("b_real:", b_real)
print("b_imag:", b_imag)
print("c_real:", c_real)
print("c_imag:", c_imag)
print("d_real:", d_real)
print("d_imag:", d_imag)
```

```python
plt.plot(t, a_real, label='a_real')
plt.plot(t, a_imag, label='a_imag')
plt.plot(t, b_real, label='b_real')
plt.plot(t, b_imag, label='b_imag')
plt.plot(t, c_real, label='c_real')
plt.plot(t, c_imag, label='c_imag')
plt.plot(t, d_real, label='d_real')
plt.plot(t, d_imag, label='d_imag')
plt.grid(True)
plt.legend()
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.title('Time Evolution of Quantum States')
plt.show()
```

Trotter Split

```python
U_approx=expm(-0.5j*hamiltonian_X*delta_t) @ expm(-1j*hamiltonian_Y*delta_t) @ expm(-0.5j*hamiltonian_X*delta_t)
a_trot_real=[]
a_trot_imag=[]

b_trot_real=[]
b_trot_imag=[]

c_trot_real=[]
c_trot_imag=[]

d_trot_real=[]
d_trot_imag=[]
```

```python
for i in range(len(t_trot)):
    result = np.linalg.matrix_power(U_approx, i) @ input_v
    a_trot_real.append(np.real(result[0, 0]))
    a_trot_imag.append(np.imag(result[0, 0]))
    
    b_trot_real.append(np.real(result[1, 0]))
    b_trot_imag.append(np.imag(result[1, 0]))
    
    c_trot_real.append(np.real(result[2, 0]))
    c_trot_imag.append(np.imag(result[2, 0]))
    
    d_trot_real.append(np.real(result[3, 0]))
    d_trot_imag.append(np.imag(result[3, 0]))
```

```python
print("a_trot_real:", a_trot_real)
print("a_trot_imag:", a_trot_imag)
print("b_trot_real:", b_trot_real)
print("b_trot_imag:", b_trot_imag)
print("c_trot_real:", c_trot_real)
print("c_trot_imag:", c_trot_imag)
print("d_trot_real:", d_trot_real)
print("d_trot_imag:", d_trot_imag)
```

```python
plt.plot(t_trot, a_trot_real, label='a_trot_real')
plt.plot(t_trot, a_trot_imag, label='a_trot_imag')
plt.plot(t_trot, b_trot_real, label='b_trot_real')
plt.plot(t_trot, b_trot_imag, label='b_trot_imag')
plt.plot(t_trot, c_trot_real, label='c_trot_real')
plt.plot(t_trot, c_trot_imag, label='c_trot_imag')
plt.plot(t_trot, d_trot_real, label='d_trot_real')
plt.plot(t_trot, d_trot_imag, label='d_trot_imag')
plt.grid(True)
plt.legend()
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.title('Trotterized Time Evolution of Quantum States')
plt.show()
```
