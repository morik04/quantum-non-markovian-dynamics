Hamiltonian Evolution

```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.linalg import expm
```

```python
delta_t=0.25
t=np.arange(0, 10, 0.01)
t_trot=np.arange(0, 10, delta_t)
```

```python
J=1  #coupling strength
X=np.array([[0, 1], [1, 0]])
Y=np.array(([0, -1j], [1j, 0]))
input_v=np.array([[0], [1], [0], [0]])
hamiltonian_X=J*0.5*np.kron(X, X)
hamiltonian_Y=J*0.5*np.kron(Y, Y)
hamiltonian=hamiltonian_X+hamiltonian_Y
```

```python
print("Hamiltonian:\n", hamiltonian)
```

```python
H=hamiltonian
```

```python
a_real=[]
a_imag=[]

b_real=[]
b_imag=[]

c_real=[]
c_imag=[]

d_real=[]
d_imag=[]
```

```python
for i in range(len(t)):
    U=expm(-1j*hamiltonian*t[i])@input_v
    a_real.append(np.real(U[0, 0]))
    a_imag.append(np.imag(U[0, 0]))
    
    b_real.append(np.real(U[1, 0]))
    b_imag.append(np.imag(U[1, 0]))
    
    c_real.append(np.real(U[2, 0]))
    c_imag.append(np.imag(U[2, 0]))
    
    d_real.append(np.real(U[3, 0]))
    d_imag.append(np.imag(U[3, 0]))
```

```python
print("a_real:", a_real)
print("a_imag:", a_imag)
print("b_real:", b_real)
print("b_imag:", b_imag)
print("c_real:", c_real)
print("c_imag:", c_imag)
print("d_real:", d_real)
print("d_imag:", d_imag)
```

```python
plt.plot(t, a_real, label='a_real')
plt.plot(t, a_imag, label='a_imag')
plt.plot(t, b_real, label='b_real')
plt.plot(t, b_imag, label='b_imag')
plt.plot(t, c_real, label='c_real')
plt.plot(t, c_imag, label='c_imag')
plt.plot(t, d_real, label='d_real')
plt.plot(t, d_imag, label='d_imag')
plt.grid(True)
plt.legend()
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.title('Time Evolution of Quantum States')
plt.show()
```

Trotter Split

```python
U_approx=expm(-0.5j*hamiltonian_X*delta_t) @ expm(-1j*hamiltonian_Y*delta_t) @ expm(-0.5j*hamiltonian_X*delta_t)
a_trot_real=[]
a_trot_imag=[]

b_trot_real=[]
b_trot_imag=[]

c_trot_real=[]
c_trot_imag=[]

d_trot_real=[]
d_trot_imag=[]
```

```python
for i in range(len(t_trot)):
    result = np.linalg.matrix_power(U_approx, i) @ input_v
    a_trot_real.append(np.real(result[0, 0]))
    a_trot_imag.append(np.imag(result[0, 0]))
    
    b_trot_real.append(np.real(result[1, 0]))
    b_trot_imag.append(np.imag(result[1, 0]))
    
    c_trot_real.append(np.real(result[2, 0]))
    c_trot_imag.append(np.imag(result[2, 0]))
    
    d_trot_real.append(np.real(result[3, 0]))
    d_trot_imag.append(np.imag(result[3, 0]))
```

```python
print("a_trot_real:", a_trot_real)
print("a_trot_imag:", a_trot_imag)
print("b_trot_real:", b_trot_real)
print("b_trot_imag:", b_trot_imag)
print("c_trot_real:", c_trot_real)
print("c_trot_imag:", c_trot_imag)
print("d_trot_real:", d_trot_real)
print("d_trot_imag:", d_trot_imag)
```

```python
plt.plot(t_trot, a_trot_real, label='a_trot_real')
plt.plot(t_trot, a_trot_imag, label='a_trot_imag')
plt.plot(t_trot, b_trot_real, label='b_trot_real')
plt.plot(t_trot, b_trot_imag, label='b_trot_imag')
plt.plot(t_trot, c_trot_real, label='c_trot_real')
plt.plot(t_trot, c_trot_imag, label='c_trot_imag')
plt.plot(t_trot, d_trot_real, label='d_trot_real')
plt.plot(t_trot, d_trot_imag, label='d_trot_imag')
plt.grid(True)
plt.legend()
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.title('Trotterized Time Evolution of Quantum States')
plt.show()
```

First Circuit

```python
from qiskit.circuit.library import UnitaryGate
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, state_fidelity
from qiskit_aer import *
from qiskit import transpile
from qiskit import ClassicalRegister
from qiskit import QuantumRegister
```

```python
Ux=UnitaryGate(expm(-0.5j*hamiltonian_X*delta_t), label='Ux')
Uy=UnitaryGate(expm(-1j*hamiltonian_Y*delta_t), label='Uy' )
n_steps=int(10/delta_t)
```

```python
shots=10000
stepwise_results = []
simulator = Aer.get_backend('qasm_simulator')

for step in range(1, n_steps + 1):
    qreg = QuantumRegister(2)
    creg = ClassicalRegister(2)
    qc1 = QuantumCircuit(qreg, creg)

    qc1.x(qreg[0])  
    for i in range(step):
        qc1.append(Ux, [qreg[0], qreg[1]])
        qc1.append(Uy, [qreg[0], qreg[1]])
        qc1.append(Ux, [qreg[0], qreg[1]])

    qc1.measure(qreg[0], creg[0])
    qc1.measure(qreg[1], creg[1])

    # Transpile and simulate
    qc_t1 = transpile(qc1, simulator)
    job = simulator.run(qc_t1, shots=shots)
    result = job.result()
    counts = result.get_counts()
    stepwise_results.append(counts)

    print(f"Step {step} counts: {counts}")
```

```python
p_01 = []
p_10 = []

for step, counts in enumerate(stepwise_results, 1):
    total = counts.get('01', 0) + counts.get('10', 0)
    if total > 0:
        prob_01 = counts.get('01', 0) / total
        prob_10 = counts.get('10', 0) / total
    else:
        prob_01 = 0
        prob_10 = 0

    p_01.append(prob_01)
    p_10.append(prob_10)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(p_01) + 1), p_01, marker='o', label="P('01')")
plt.plot(range(1, len(p_10) + 1), p_10, marker='s', label="P('10')")
plt.title("Probability of 01 and 10")
plt.xlabel("Step")
plt.ylabel("Probability")
plt.ylim(0, 1)
plt.grid(True)
plt.legend()
plt.show()
```

Second Circuit

```python
Z =2*np.array([[1, 0], [0, -1]])  # Pauli-Z
Z_component=UnitaryGate(expm(-1j*J*delta_t*0.25*Z),label=r'$e^{-i 0.25 J \delta t \, Z}$')
```

```python
shots = 10000
stepwise_results=[]
simulator = Aer.get_backend('qasm_simulator')

for step in range(1, n_steps + 1):
    qc2 = QuantumCircuit(3, 3)

    qc2.x(0)  # Apply X gate to qubit 0 only once at the start

    for i in range(step):
        qc2.h(0)
        qc2.h(1)

        qc2.cx(0, 2)
        qc2.cx(1, 2)
        qc2.append(Z_component, [2])
        qc2.cx(1, 2)
        qc2.cx(0, 2)

        qc2.h(0)
        qc2.h(1)

    qc2.measure([0, 1, 2], [0, 1, 2])

    job=simulator.run(transpile(qc2, simulator), shots=shots)
    result=job.result()
    counts=result.get_counts()
    stepwise_results.append(counts)

    print(f"Step {step} counts: {counts}")

```

```python
p_010 = []
p_001 = []

for step, counts in enumerate(stepwise_results, 1):
    total = counts.get('010', 0) + counts.get('001', 0)
    if total > 0:
        prob_010 = counts.get('010', 0) / total
        prob_001 = counts.get('001', 0) / total
    else:
        prob_010 = 0
        prob_001 = 0

    p_010.append(prob_010)
    p_001.append(prob_001)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(p_010) + 1), p_010, marker='o', label="P('010')")
plt.plot(range(1, len(p_001) + 1), p_001, marker='s', label="P('001')")
plt.title("Probability of 010 and 001 Over Steps")
plt.xlabel("Step")
plt.ylabel("Probability")
plt.ylim(0, 1)
plt.grid(True)
plt.legend()
plt.show()
```
